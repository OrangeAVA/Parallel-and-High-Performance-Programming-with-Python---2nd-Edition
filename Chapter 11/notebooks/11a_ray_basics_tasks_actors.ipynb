# Ray basics: init, Tasks & Actors

```

# Se necessario: pip install ray
import ray, os
ray.shutdown()
ray.init()  # locale; per cluster: ray.init(address="auto")
print("Dashboard:", "http://127.0.0.1:8265")
resources = ray.available_resources()
print("Resources:", resources)
```

```

import time

@ray.remote
def compute_square(x): 
    return x * x

futures = [compute_square.remote(i) for i in range(10)]
print("Squares:", ray.get(futures))
```

```

@ray.remote
def add(x,y): return x+y

@ray.remote
def multiply(x,y): return x*y

result = multiply.remote(add.remote(1,2), 3)
print("compose:", ray.get(result))
```

```

@ray.remote
class Counter:
    def __init__(self): self.count = 0
    def increment(self):
        self.count += 1
        return self.count

c1 = Counter.remote()
c2 = Counter.remote()
_ = ray.get([c1.increment.remote(), c1.increment.remote()])
res = ray.get([c2.increment.remote(), c1.increment.remote()])  # [1,3]
print("Actors:", res)
```
