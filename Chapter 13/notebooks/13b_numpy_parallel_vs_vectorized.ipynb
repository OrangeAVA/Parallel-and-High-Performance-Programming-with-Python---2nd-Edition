```
import numpy as np
from joblib import Parallel, delayed
import time

arr=np.random.rand(200_000)
mean, std = arr.mean(), arr.std()

start=time.time();
res_joblib = Parallel(n_jobs=-1)(delayed(lambda v:(v-mean)/std)(v) for v in arr);
print('joblib', round(time.time()-start,3),'s')

start=time.time();
res_vec = (arr-mean)/std
print('vectorized', round(time.time()-start,6),'s')
```
