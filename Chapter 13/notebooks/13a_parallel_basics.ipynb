```
from joblib import Parallel, delayed

def process_number(n):
    return n*n

Parallel(n_jobs=4)(delayed(process_number)(i) for i in range(10))
```

```
from joblib import Parallel, delayed
import time

def expensive_operation(n):
    time.sleep(1.0)
    return n**3

# Parallel vs sequential
start=time.time(); res=Parallel(n_jobs=3)(delayed(expensive_operation)(i) for i in range(6)); print(res, 'parallel', round(time.time()-start,2),'s')
start=time.time(); res=[expensive_operation(i) for i in range(6)]; print(res, 'sequential', round(time.time()-start,2),'s')
```
