```
from joblib import Memory
import time, os
memory=Memory(location='../cache_dir', verbose=0)

@memory.cache
def f(x):
    time.sleep(0.5)
    return x*x

start=time.time(); print(f(10)); print('first', round(time.time()-start,3),'s')
start=time.time(); print(f(10)); print('cached', round(time.time()-start,3),'s')
```
