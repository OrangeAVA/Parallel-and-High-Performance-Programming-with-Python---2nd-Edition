# 08c — Groupby on `prices_sample.csv`

Mimics the Kaggle example: read CSV → clean columns/dtypes → groupby on `[symbol, year]`.

```
import dask.dataframe as dd
import pandas as pd
from pathlib import Path
DATA = Path('../data')
```

```
# Dask read (with dtype hint for volume and parse_dates)
ddf = dd.read_csv(DATA / 'prices_sample.csv', dtype={'volume':'float64'}, parse_dates=['date'])
# Drop CSV index column if present
ddf = ddf.drop(columns=['Unnamed: 0'], errors='ignore')

ddf.head()
```

```
ddf.npartitions
```

```
# Groupby: mean by symbol and year
mean_dd = ddf.groupby(['symbol', ddf['date'].dt.year])[['open','high','low','close','volume']].mean()
%time result_dd = mean_dd.compute()
result_dd.head(10)
```

### Compare with pandas (single-machine)

```
df = pd.read_csv(DATA / 'prices_sample.csv', dtype={'volume':'float64'}, parse_dates=['date'])
df = df.drop(columns=['Unnamed: 0'], errors='ignore')
%time result_pd = df.groupby(['symbol', df['date'].dt.year])[['open','high','low','close','volume']].mean()
result_pd.head(10)
```
